```python
#calculator
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

def mul(a,b):
    return a*b

def div(a,b):
    return a/b

def cal():
    a = int(input("enter a ="))
    b = int(input("enter b="))
    o = input("enter + - * / ")
    
    if o == "+":
        print(add(a,b))
    elif o== "-":
        print(sub(a,b))
    elif o== "*":
        print(mul(a,b))
    elif o== "/":
        print(div(a,b))
    else:
        print("invalid operator")
    
cal()
```
